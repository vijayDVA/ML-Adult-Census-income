```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df[df == '?'] = np.nan
```

```python
for col in ['workclass', 'occupation', 'native.country']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.drop(['income'], axis=1)

y = df['income']
```

```python
from sklearn import preprocessing

categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
for feature in categorical:
        le = preprocessing.LabelEncoder()
        X[feature] = le.fit_transform(X[feature])
        
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X= pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_x= pca.fit_transform(X)
pca.explained_variance_ratio_
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print('Logistic Regression accuracy score with all the features: {0:0.4f} %'. format(accuracy_score(y_test, y_pred)*100))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knncla = KNeighborsClassifier(n_neighbors=5,n_jobs=-1)
knncla.fit(X_train, y_train)
y_pred = knncla.predict(X_test)

print('KNN accuracy score with all the features: {0:0.4f} %'. format(accuracy_score(y_test, y_pred)*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfcla = RandomForestClassifier(n_estimators=100,random_state=9,n_jobs=-1)
rfcla.fit(X_train, y_train)
y_pred = rfcla.predict(X_test)

print('Random Forest accuracy score with all the features: {0:0.4f} %'. format(accuracy_score(y_test, y_pred)*100))
```
